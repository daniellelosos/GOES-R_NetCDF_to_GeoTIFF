```python
import numpy as np
import s3fs
import os
import os.path
import netCDF4
import matplotlib.pyplot as plt
import xarray as xr
import rasterio
from rasterio.mask import mask
import geopandas as gpd
```

```python
# Adjust the variable inputs to create the prefix for the file of interest
bucket_name = 'noaa-goes16' # Change to 'noaa-goes17' for GOES-17 satellite
product_name = 'ABI-L2-CMIPC'# Choose a GOES-R L2+ product with a binary variable, 
                            #    ie. Clear Sky Mask (ACMC/ACMF) or Aersol Detection (ADPC/ADPF)
year = 2020 
day_of_year = 296 
hour = 16

# Adjust pathname to local directory
directory = "/Users/daniellelosos/GOESforGIS/"
```

```python
# Make new directories to store files
if not os.path.isdir(directory + product_name +"_netCDFs"):
    os.mkdir(directory + product_name +"_netCDFs")
if not os.path.isdir(directory + product_name +"_Clipped_netCDFs"):
    os.mkdir(directory + product_name +"_Clipped_netCDFs")
```

```python
# Use anonymous credentials to access public data  from AWS
fs = s3fs.S3FileSystem(anon=True)

# Write prefix for the files of interest, and list all files beginning with this prefix.
prefix = f'{bucket_name}/{product_name}/{year}/{day_of_year:03.0f}/{hour:02.0f}/'
print(prefix)
files = fs.ls(prefix)
```

```python
#Read the first netCDF file in the list (closest to the top of the hour) and download
first_file = files[0]
path = directory + product_name +"_netCDFs/" + first_file.split("/")[-1]
fs.download(first_file, path)
```

```python
f = netCDF4.Dataset(path)

# Choose the first variable/subdataset in the netCDF file to convert to a GeoTIFF.
# To select altnative variable stored in the file, adjust the list number, ie. [0] selects the 1st variable 
var = list(f.variables.keys())[0] 
print ("Selected variable: " + var)

# Uncomment the line below to reveal the list of variables stored in the file
#print(list(f.variables.keys()))
```

```python
# Open dataset and parse variable information
ds = xr.open_dataset(path)
var_name = ds[var].long_name
units_name = ds[var].units
variable = ds[var].data
```

```python
# Plot the variable to check if data loaded correctly
fig = plt.figure(figsize=(4,4),dpi=200)
im = plt.imshow(variable, cmap='Greys_r')
cb = fig.colorbar(im, orientation='horizontal')
cb.set_label(var_name + '('+ units_name +')')
plt.show()
```

```python
# Open netCDF file with rasterio
netCDF_file = rasterio.open('netcdf:{0}:{1}'.format(path, var))

# Open shapefile with geopandas and view attributes
# Download the state border shapefile from:
#'https://www.census.gov/geographies/mapping-files/time-series/geo/carto-boundary-file.html'
us_state_borders = gpd.read_file('/Users/daniellelosos/GOESforGIS/cb_2018_us_state_5m/cb_2018_us_state_5m.shp')
us_state_borders
```

```python
# If desired, select geometry subset by editing the field and value
state_border = us_state_borders[us_state_borders['NAME'] == 'Michigan']
```

```python
# Parse projection information from netCDF file's metadata
sat_h = ds['goes_imager_projection'].perspective_point_height
lon_origin = ds['goes_imager_projection'].longitude_of_projection_origin

# Write proj-4 string for GOES-R geostationary projection
proj_string = f'+proj=geos +h={sat_h} +lon_0={lon_origin} +sweep=x'
print(proj_string)
```

```python
# Reproject shapefile to match the GOES-R geostationary projection
reprojected_border = state_border.to_crs(crs=proj_string)
```

```python
# Pass shapefile (a GeoDataFrame) through a function to parse the geometry of the feature into coordinate values 
# formatted to be readable by rasterio.
def getFeatures(gdf):
    import json
    return [json.loads(gdf.to_json())['features'][0]['geometry']]

coords = getFeatures(reprojected_border)
```

```python
# Clip netCDF image by masking values outside of the shapefile's boundary
out_img, out_transform = mask(netCDF_file, shapes=coords, crop=True)
```

```python
# Copy the shapefile's metadata and update with the clipped image's metadata
out_meta = netCDF_file.meta.copy()
out_meta.update({"driver": "GTiff",
                 "height": out_img.shape[1],
                 "width": out_img.shape[2],
                 "transform": out_transform,
                 "crs": proj_string})
```

```python
# Set output pathname and write clipped GeoTIFF to the path
clipped_name = product_name + "_" + var + ".tif"
geotiff_pathname = os.path.join(directory + product_name +"_Clipped_netCDFs", clipped_name)

with rasterio.open(geotiff_pathname, "w", **out_meta) as output:
     output.write(out_img)
```

```python
# Check to see if clipped GeoTIFF appears as expected by visualizing image
with rasterio.open(geotiff_pathname) as img:
    data = img.read(1)

fig = plt.figure(figsize=(4,4),dpi=200)
im = plt.imshow(np.flipud(data), cmap='Greys_r')
cb = fig.colorbar(im, orientation='horizontal')
cb.set_label(var_name + '('+ units_name +')')
plt.axis('off')
plt.show()
```
